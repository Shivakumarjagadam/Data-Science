# Matplotlib
Matplotlib is one of the most popular data visualization libraries in Python. It 
provides a wide range of tools to create high-quality charts and graphs, making 
it an essential tool for data scientists and analysts. One of the key components 
of Matplotlib is pyplot, which provides a simple interface for creating plots.

```python
from matplotlib import pyplot as plt
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5]
# y=[1,2,3,4,5]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x)
plt.title("line graph")

plt.show()
```

```python
from matplotlib import pyplot as pyplot

xpoints=[3,7,9,15,8,3]
ypoints=[4,6,18,25,12,10]
zpoints=[2,4,6,8,15,11]
plt.plot(xpoints,color="red",linestyle="dashed",linewidth="2.2")
plt.plot(ypoints,color="blue",linestyle="dotted",linewidth="3.5")
plt.plot(zpoints,color="black",linewidth="5")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("multiple lines")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
# x as 3*pi and y as sin wave
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

#plot by matplotlib
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(0,3*np.pi,0.1)

y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)

plt.subplot(2,1,2)
plt.plot(x,y_cos)

plt.show()
```

```python
from matplotlib import pyplot as plt

branches=["CSE","ECE","EEE","MECH","CIVIL"]
students=[240,240,140,140,50]
c=["red","blue","green","yellow","grey"]
plt.title("Branches")

plt.pie(students,labels=branches,colors=c,shadow=True,explode=[0.1,0,0,0.2,0],autopct='%1.f%%')
plt.legend(branches,bbox_to_anchor=(0,1),title="Branches")
plt.show()
```
