# Matplotlib

Matplot is one of the most popular data visualization libraries in python.it provides a wide range of tools to create high-quality charts and graphs.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create a line plot
plt.plot(x, y, marker='o')

# Add title and labels
plt.title('Line Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fruits=['apple','banana','cherry','mango']
count=[10,30,80,100]
c=['red','orange','yellow','green']
plt.pie(count,colors=c,explode=[0.1,0.1,0.1,0.1],shadow=True,labels=fruits)
plt.legend(fruits)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,6]

plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("its a line graph")

plt.plot(x,y)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
xpoints=[2,4,6,8,10,12]
plt.plot(xpoints,linestyle="dashed",linewidth="4")

ypoints=[1,3,5,17,9,11]
plt.plot(ypoints,linestyle="dashed",linewidth="2")
plt.title("2-line graph")
plt.xlabel("X-axis")

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x=[1,1,2,2.5,3]
y=[1,3,5,9,12]

s=[100,200,300,400]
plt.scatter(x,y,color="red",sizes=s)
plt.title("SCATTER")

x=[1,3,5,7,9]
y=[1,2,3,5,7]
plt.scatter(x,y,color="blue",sizes=s)

plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

subject=["DAA","DSA","WEBDEV","JAVA","OOPS","PYTHON"]
values=[75,80,65,89,80,100]
c=["red","green","blue","yellow","grey","black"]

plt.bar(subject,values,color=c,width=0.5)

plt.xlabel("Subjects")
plt.ylabel("Values")
plt.title("PERFORMANCE")

plt.show()
```
