```python
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(0,3*np.pi,0.1)
labels=["Data1","Data2"]

y_sin=np.sin(x)

y_cos=np.cos(x)

plt.xlabel("X")
plt.ylabel("Data values")

plt.title("scatter plot of Data")

# plt.subplot(2,1,1)

plt.scatter(x,y_sin,color="blue",marker="o")

plt.scatter(x,y_cos,color="red",marker="s")


plt.legend(labels,bbox_to_anchor=(0.8,1))
plt.show()






x=["Category A","Category B","Category C","Category D","Category E"]

y=[18,30,90,35,45]

label=["Data 1"]

plt.title("Bar Plot of Different Data")
plt.xlabel("X")
plt.ylabel("Data values")

# plt.subplot(2,1,2)

plt.bar(x,y,color="green")
plt.legend(label,bbox_to_anchor=(1,1))

plt.show()
```
