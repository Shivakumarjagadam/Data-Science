```python
import pandas as pd

mydata={
  "Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct"],
  
  "Temperature":[30,32,45,50,60,70,75,80,65,55],
  
  "Season":["Winter","Winter","Spring","Spring","Spring","Summer","Summer","Summer","Fall","Fall"]
}

df=pd.DataFrame(mydata)
print("1 :")
print(df)

print("\n 2.\n")
mean_temp=df["Temperature"].mean()
print("\n","the mean_temp:",mean_temp)


mid_temp=df["Temperature"].median()
print("\n","the mid_temp:",mid_temp)

mod_season=df["Season"].mode()
print("\n the mod_season: ",mod_season)




```

```python
import pandas as pd

grouped = df.groupby('Season')

grp_temp = grouped['Temperature'].mean()

print(" \n avg temperature grouped by Seasons :", grp_temp)
```
